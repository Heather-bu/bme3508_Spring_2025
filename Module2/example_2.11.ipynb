# Example 2.11: Fourier Transform of a Triangular Pulse

Let's derive the Fourier Transform of a triangular pulse function defined as:

$$x(t) = \text{tri}\left(\frac{t}{\tau}\right) = \begin{cases}
1 - \frac{|t|}{\tau}, & |t| < \tau \\
0, & |t| \geq \tau
\end{cases}$$

The Fourier Transform is given by:

$$X(\omega) = \int_{-\infty}^{\infty} x(t) e^{-j\omega t} dt$$

For our triangular pulse, we can write:

$$X(\omega) = \int_{-\tau}^{\tau} \left(1 - \frac{|t|}{\tau}\right) e^{-j\omega t} dt$$

Let's split this integral into two parts:

$$X(\omega) = \int_{-\tau}^{0} \left(1 + \frac{t}{\tau}\right) e^{-j\omega t} dt + \int_{0}^{\tau} \left(1 - \frac{t}{\tau}\right) e^{-j\omega t} dt$$

For the first integral:
$$I_1 = \int_{-\tau}^{0} \left(1 + \frac{t}{\tau}\right) e^{-j\omega t} dt$$

Let's split this into two parts:
$$I_1 = \int_{-\tau}^{0} e^{-j\omega t} dt + \frac{1}{\tau}\int_{-\tau}^{0} t e^{-j\omega t} dt$$

For the first part:
$$\int_{-\tau}^{0} e^{-j\omega t} dt = \left[ \frac{e^{-j\omega t}}{-j\omega} \right]_{-\tau}^{0} = \frac{1}{-j\omega}(1 - e^{j\omega\tau}) = \frac{1}{j\omega}(e^{j\omega\tau} - 1)$$

For the second part, we use integration by parts with $u = t$ and $dv = e^{-j\omega t}dt$:
$$\int_{-\tau}^{0} t e^{-j\omega t} dt = \left[ t \cdot \frac{e^{-j\omega t}}{-j\omega} \right]_{-\tau}^{0} - \int_{-\tau}^{0} \frac{e^{-j\omega t}}{-j\omega} dt$$

$$= \left[ \frac{-t e^{-j\omega t}}{j\omega} \right]_{-\tau}^{0} - \frac{1}{-j\omega}\int_{-\tau}^{0} e^{-j\omega t} dt$$

$$= \frac{-0 \cdot 1}{j\omega} - \frac{-(-\tau) \cdot e^{j\omega\tau}}{j\omega} - \frac{1}{-j\omega} \cdot \frac{1}{-j\omega}(1 - e^{j\omega\tau})$$

$$= \frac{\tau e^{j\omega\tau}}{j\omega} - \frac{1}{(j\omega)^2}(e^{j\omega\tau} - 1)$$

Therefore:
$$I_1 = \frac{1}{j\omega}(e^{j\omega\tau} - 1) + \frac{1}{\tau} \left[ \frac{\tau e^{j\omega\tau}}{j\omega} - \frac{1}{(j\omega)^2}(e^{j\omega\tau} - 1) \right]$$

$$= \frac{1}{j\omega}(e^{j\omega\tau} - 1) + \frac{e^{j\omega\tau}}{j\omega} - \frac{1}{\tau(j\omega)^2}(e^{j\omega\tau} - 1)$$

For the second integral:
$$I_2 = \int_{0}^{\tau} \left(1 - \frac{t}{\tau}\right) e^{-j\omega t} dt$$

Similarly, we split this into two parts:
$$I_2 = \int_{0}^{\tau} e^{-j\omega t} dt - \frac{1}{\tau}\int_{0}^{\tau} t e^{-j\omega t} dt$$

For the first part:
$$\int_{0}^{\tau} e^{-j\omega t} dt = \left[ \frac{e^{-j\omega t}}{-j\omega} \right]_{0}^{\tau} = \frac{1}{-j\omega}(e^{-j\omega\tau} - 1) = \frac{1}{j\omega}(1 - e^{-j\omega\tau})$$

For the second part, using integration by parts again:
$$\int_{0}^{\tau} t e^{-j\omega t} dt = \left[ t \cdot \frac{e^{-j\omega t}}{-j\omega} \right]_{0}^{\tau} - \int_{0}^{\tau} \frac{e^{-j\omega t}}{-j\omega} dt$$

$$= \left[ \frac{-t e^{-j\omega t}}{j\omega} \right]_{0}^{\tau} - \frac{1}{-j\omega}\int_{0}^{\tau} e^{-j\omega t} dt$$

$$= \frac{-\tau e^{-j\omega\tau}}{j\omega} - \frac{-0 \cdot 1}{j\omega} - \frac{1}{-j\omega} \cdot \frac{1}{-j\omega}(e^{-j\omega\tau} - 1)$$

$$= \frac{-\tau e^{-j\omega\tau}}{j\omega} - \frac{1}{(j\omega)^2}(1 - e^{-j\omega\tau})$$

Therefore:
$$I_2 = \frac{1}{j\omega}(1 - e^{-j\omega\tau}) - \frac{1}{\tau} \left[ \frac{-\tau e^{-j\omega\tau}}{j\omega} - \frac{1}{(j\omega)^2}(1 - e^{-j\omega\tau}) \right]$$

$$= \frac{1}{j\omega}(1 - e^{-j\omega\tau}) + \frac{e^{-j\omega\tau}}{j\omega} - \frac{1}{\tau(j\omega)^2}(1 - e^{-j\omega\tau})$$

Solving these integrals and combining the results:

$$X(\omega) = \tau \cdot \text{sinc}^2\left(\frac{\omega\tau}{2}\right) = \tau \cdot \left(\frac{\sin(\omega\tau/2)}{\omega\tau/2}\right)^2$$

This is the Fourier Transform of the triangular pulse function.


The Significance of the Sinc Function in Fourier Transforms

The result we derived above, $X(\omega) = \tau \cdot \text{sinc}^2\left(\frac{\omega\tau}{2}\right)$, highlights the importance of the sinc function in Fourier analysis.

The sinc function, defined as $\text{sinc}(x) = \frac{\sin(x)}{x}$, appears naturally in many Fourier transform pairs:

1. **Rectangular Pulse**: The Fourier transform of a rectangular pulse is a sinc function. This is the dual of our result, where a triangular pulse transforms to a sinc-squared function.

2. **Sampling and Interpolation**: The sinc function is the ideal interpolation function in signal processing. When we sample a continuous signal, the reconstruction involves sinc interpolation.

3. **Spectral Characteristics**: The sinc function has zeros at integer multiples of Ï€ (except at the origin), which creates frequency nulls in systems using rectangular windowing.

4. **Resolution and Bandwidth**: The width of the main lobe of the sinc function is inversely proportional to the width of the time-domain signal, illustrating the time-bandwidth product principle.

5. **Filter Design**: Ideal lowpass filters have sinc impulse responses, though they're not physically realizable due to their infinite extent.

In our triangular pulse example, the sinc-squared spectrum indicates smoother frequency roll-off compared to a rectangular pulse (which has a simple sinc spectrum). This demonstrates how the smoothness of a function in the time domain relates to the decay rate of its spectrum in the frequency domain.
