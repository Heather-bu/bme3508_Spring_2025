In this example, we will explore the periodic rectangular pulse function.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define signal parameters
A = 1  # Amplitude
T = 10 * np.pi  # Period
tau = 2 * np.pi  # Pulse width
omega_0 = 2 * np.pi / T  # Fundamental frequency

# Define time domain
t = np.linspace(-2 * T, 2 * T, 1000)

# Define the periodic rectangular pulse function
def rectangular_pulse(t, A, tau, T):
    return A * np.where(np.mod(t + tau/2, T) < tau, 1, 0)

# Compute the signal values
x_t = rectangular_pulse(t, A, tau, T)

# Plot the time-domain signal
plt.figure(figsize=(10, 4))
plt.plot(t, x_t, label=r"$x(t)$", linewidth=2)
plt.xlabel(r"$t$")
plt.ylabel(r"$x(t)$")
plt.title("Periodic Rectangular Pulse Function")

# Modify x-axis ticks to be in terms of Ï€
xticks_positions = np.arange(-2*T, 2*T + np.pi, 2*np.pi)
xticks_labels = [rf"${int(round(x/np.pi))}\pi$" if x != 0 else r"$0$" for x in xticks_positions]
plt.xticks(xticks_positions, xticks_labels)

plt.grid()
plt.legend()
plt.show()
```

The given signal $x(t)$ appears to be a periodic rectangular pulse train with period $T = 10\pi$. To analyze the magnitude and phase spectrum of the signal, we derive its Fourier series representation.

## Step 1: Define the Signal Mathematically
From the image, we observe that $x(t)$ consists of a repeating rectangular pulse of amplitude $A$, width $2\pi$, and period $T = 10\pi$. A single pulse centered at $t = 0$ can be written as:

$$
x(t) =
\begin{cases}
A, & -\pi \leq t \leq \pi \\
0, & \pi < |t| \leq 9\pi
\end{cases}
$$

Since the function repeats every $T = 10\pi$, we express $x(t)$ using its Fourier series.

---


## Step 2: Compute the Fourier Series Coefficients
The Fourier series of a periodic function is given by:

$$
x(t) = \sum_{k=-\infty}^{\infty} c_k e^{j k \omega_0 t}
$$

where

$$
c_k = \frac{1}{T} \int_{T} x(t) e^{-j k \omega_0 t} dt
$$

The fundamental angular frequency is:

$$
\omega_0 = \frac{2\pi}{T} = \frac{2\pi}{10\pi} = \frac{1}{5}
$$

Since $x(t)$ is nonzero only in the interval $-\pi \leq t \leq \pi$, the integral simplifies to:

$$
c_k = \frac{1}{10\pi} \int_{-\pi}^{\pi} A e^{-j k \frac{1}{5} t} dt
$$

Evaluating the integral:

$$
c_k = \frac{A}{10\pi} \int_{-\pi}^{\pi} e^{-j k \frac{1}{5} t} dt
$$

Using the standard integral result:

$$
\int e^{-j at} dt = \frac{e^{-j at}}{-j a}
$$

we obtain:

$$
c_k = \frac{A}{10\pi} \left[ \frac{e^{-j k \frac{1}{5} t}}{-j k \frac{1}{5}} \right]_{-\pi}^{\pi}
$$

$$
= \frac{A}{10\pi} \left[ \frac{e^{-j k \frac{1}{5} \pi} - e^{j k \frac{1}{5} \pi}}{-j k \frac{1}{5}} \right]
$$

$$
= \frac{A}{10\pi} \left[ \frac{-2j \sin(k \frac{1}{5} \pi)}{-j k \frac{1}{5}} \right]
$$

$$
= \frac{A}{10\pi} \times \frac{2 \sin(k \frac{1}{5} \pi)}{k \frac{1}{5}}
$$

$$
= \frac{A}{10\pi} \times \frac{10 \sin(k \frac{1}{5} \pi)}{k}
$$

$$
= \frac{A}{\pi} \frac{\sin(k \frac{1}{5} \pi)}{k}
$$

For $k = 0$, we compute $c_0$ separately:

$$
c_0 = \frac{1}{10\pi} \int_{-\pi}^{\pi} A dt = \frac{A}{10\pi} \times 2\pi = \frac{2A}{10} = \frac{A}{5}
$$

---

## Step 3: Compute Magnitude and Phase Spectra
- **Magnitude Spectrum:**

$$
|c_k| = \left| \frac{A}{\pi} \frac{\sin(k \frac{1}{5} \pi)}{k} \right|
$$

- **Phase Spectrum:**

$$
\angle c_k = \angle \left( \frac{\sin(k \frac{1}{5} \pi)}{k} \right)
$$

Since $\sin(x)$ is odd,

- For $k > 0$, $\angle c_k = 0$ if $\sin(k \frac{1}{5} \pi) > 0$ and $\pi$ if $\sin(k \frac{1}{5} \pi) < 0$.
- For $k < 0$, $\angle c_k = \pi$ if $\sin(k \frac{1}{5} \pi) > 0$ and $0$ if $\sin(k \frac{1}{5} \pi) < 0$.

---

### Deriving the first five negative and positive Fourier coefficients:

For $k = 0$:
$c_0 = A/5 = 0.2A$

For $k = 1$:
$c_1 = (A/\pi) \cdot (\sin(\pi/5) / 1) = (A/\pi) \cdot \sin(\pi/5) \approx 0.1876A$

For $k = 2$:
$c_2 = (A/2\pi) \cdot (\sin(2\pi/5)) = (A/2\pi) \cdot \sin(2\pi/5) \approx 0.151$

For $k = 3$:
$c_3 = (A/\pi) \cdot (\sin(3\pi/5) / 3) = (A/\pi) \cdot \sin(3\pi/5) / 3 \approx 0.0957A$

$c_4 = (A/\pi) \cdot (\sin(4\pi/5) / 4) = (A/\pi) \cdot \sin(4\pi/5) / 4 \approx 0.0303A$

For $k = 5$:
$c_5 = (A/\pi) \cdot (\sin(5\pi/5) / 5) = (A/\pi) \cdot \sin(\pi) / 5 = 0$

For $k = -1$:
$c_{-1} = (A/\pi) \cdot (\sin(-\pi/5) / -1) = (A/\pi) \cdot \sin(\pi/5) \approx 0.1876A$

For $k = -2$:
$c_{-2} = (A/\pi) \cdot (\sin(-2\pi/5) / -2) = (A/\pi) \cdot \sin(2\pi/5) / 2 \approx 0.151$

For $k = -3$:
$c_{-3} = (A/\pi) \cdot (\sin(-3\pi/5) / -3) = (A/\pi) \cdot \sin(3\pi/5) / 3 \approx 0.0957A$

For $k = -4$:
$c_{-4} = (A/\pi) \cdot (\sin(-4\pi/5) / -4) = (A/\pi) \cdot \sin(4\pi/5) / 4 \approx 0.0303A$

For $k = -5$:
$c_{-5} = (A/\pi) \cdot (\sin(-5\pi/5) / -5) = (A/\pi) \cdot \sin(-\pi) / -5 = 0$

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider, IntSlider

# Define signal parameters
A = 1  # Amplitude
tau = 2 * np.pi  # Pulse width

# Define the periodic rectangular pulse function
def rectangular_pulse(t, A, tau, T):
    return A * np.where(np.mod(t + tau/2, T) < tau, 1, 0)

# Function to update plots based on T and number of harmonics
def update_plots(T, num_harmonics):
    T = T * np.pi
    omega_0 = 2 * np.pi / T  # Fundamental frequency
    t = np.linspace(-2 * T, 2 * T, 1000)
    x_t = rectangular_pulse(t, A, tau, T)
    
    # Define Fourier series coefficients
    k_values = np.arange(-num_harmonics, num_harmonics + 1)  # Range of k values for spectrum
    c_k = (A / np.pi) * np.sinc(k_values * tau / T)  # Update sinc function with tau/T

    # Recreate the signal using the Fourier series
    x_t_recreated = np.zeros_like(t, dtype=np.complex128)  # Ensure complex type to handle complex addition
    for k in k_values:
        x_t_recreated += c_k[k + num_harmonics] * np.exp(1j * k * omega_0 * t)/1.6

    # Compute magnitude and phase
    magnitude_spectrum = np.abs(c_k)
    phase_spectrum = np.angle(c_k)

    # Create a single figure with 2x2 subplots
    fig, axs = plt.subplots(2, 2, figsize=(12, 12))

    # Plot the time-domain signal and the recreated signal
    axs[0, 0].plot(t, x_t, label=r"$x(t)$", linewidth=2)
    axs[0, 0].plot(t, x_t_recreated.real, label=r"Recreated $x(t)$", linestyle='--', linewidth=2)
    axs[0, 0].set_xlabel(r"$t$")
    axs[0, 0].set_ylabel(r"$x(t)$")
    axs[0, 0].set_title("Periodic Rectangular Pulse Function")
    xticks_positions = np.arange(-2*T, 2*T + np.pi, T/5)
    xticks_labels = [rf"${int(round(x/np.pi))}\pi$" if x != 0 else r"$0$" for x in xticks_positions]
    axs[0, 0].set_xticks(xticks_positions)
    axs[0, 0].set_xticklabels(xticks_labels)
    axs[0, 0].grid()
    axs[0, 0].legend()

    # Plot c_k values
    axs[0, 1].stem(k_values, c_k, basefmt=" ")
    axs[0, 1].set_xlabel(r"$k$")
    axs[0, 1].set_ylabel(r"$c_k$")
    axs[0, 1].set_title("Fourier Coefficients $c_k$")
    axs[0, 1].grid()

    # Plot phase spectrum
    axs[1, 0].stem(k_values, phase_spectrum, basefmt=" ")
    axs[1, 0].set_xlabel(r"$k$")
    axs[1, 0].set_ylabel(r"$\angle c_k$")
    axs[1, 0].set_title("Phase Spectrum")
    axs[1, 0].grid()
    
    # Plot magnitude spectrum
    axs[1, 1].stem(k_values, magnitude_spectrum, basefmt=" ")
    axs[1, 1].set_xlabel(r"$k$")
    axs[1, 1].set_ylabel(r"$|c_k|$")
    axs[1, 1].set_title("Magnitude Spectrum")
    axs[1, 1].grid()

    # Adjust layout
    plt.tight_layout()
    plt.show()

# Create interactive sliders for T and number of harmonics
interact(update_plots, 
         T=FloatSlider(value=10, min=4, max=200, step=1, description='T'),
         num_harmonics=IntSlider(value=20, min=1, max=50, step=1, description='Harmonics'))
```
